```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df=pd.read_csv("iris_dataset.csv")
df
```

```python
# df1=df[['sepal length (cm)','sepal width (cm)','target','petal length (cm)','petal width (cm)']]
# df1
```

```python
df1=df[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
df1
```

```python
P_test=df.iloc[149]
P_test
```

```python
df2=df1.drop(df1.index[149])
df2
```

```python
X=P_test.iloc[0]
Y=P_test.iloc[1]
Z=P_test.iloc[2]
T=P_test.iloc[3]
(X,Y,Z,T)
```

```python
list_Euclidean_Distance=[]
for col in range(len(df2)):
   list_Euclidean_Distance.append(math.sqrt( ((X-df2.iloc[col,0])**2)+((Y-df2.iloc[col,1])**2)+((Z-df2.iloc[col,2])**2)+((T-df2.iloc[col,3])**2)))
list_Euclidean_Distance
```

```python
df3=pd.DataFrame(list_Euclidean_Distance)
df3.columns=['Euclidean Distance']
df3
```

```python
idx = np.argsort(list_Euclidean_Distance)[:3]
idx
```

```python
np.array(list_Euclidean_Distance)[idx]
```

```python
list_target=[]
list_target=df["target"][idx]
list_target
```
