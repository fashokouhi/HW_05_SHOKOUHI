```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris_dataset.csv')
df
```

```python
setosa_df = df[df['target']==0].drop(columns=['target'])
versicolor_df=df[df['target']==1].drop(columns=['target'])
virginica_df=df[df['target']==2].drop(columns=['target'])
```

```python
setosa_maen=setosa_df.mean()
setosa_var=setosa_df.var()
setosa_range=setosa_df.max()-setosa_df.min()
```

```python
setosa_chart=pd.DataFrame({
    'setosa_mean':setosa_maen,
    'setosa_var':setosa_var,
    'setosa_range':setosa_range
})
setosa_chart
```

```python
versicolor_mean=versicolor_df.mean()
versicolor_var=versicolor_df.var()
versicolor_range=versicolor_df.max()-versicolor_df.min()
```

```python
versicolor_chart=pd.DataFrame({
    'versicolor_mean':versicolor_mean,
    'versicolor_var':versicolor_var,
    'versiclolor_range':versicolor_range
})
versicolor_chart
```

```python
virginica_mean=virginica_df.mean()
virginica_var=virginica_df.var()
virginica_range=virginica_df.max()-versicolor_df.min()
```

```python
virginica_chart=pd.DataFrame({
    'virginica_mean':virginica_mean,
    'virginica_var':virginica_var,
    'virginica_range':virginica_range
})
virginica_chart
```

```python
setosa_corr=setosa_df.corr()
setosa_corr
```

```python
versicolor_corr=versicolor_df.corr()
versicolor_corr
```

```python
virginica_corr=virginica_df.corr()
virginica_corr
```

```python
sns.heatmap(setosa_corr,annot=True)


```

```python
np.fill_diagonal(setosa_corr.values, np.nan) 
list_corr_s=setosa_corr.unstack()
max_corr_s=list_corr_s.max()
max_corr_index_s=list_corr_s.idxmax()
print(f"max_correlation:{max_corr_s}\nThe highest correlation between:{max_corr_index_s[0] }and {max_corr_index_s[1]}")
```
