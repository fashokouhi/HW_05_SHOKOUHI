```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df=pd.read_csv("iris_dataset.csv")
df
```

```python
df1=df[['sepal length (cm)','sepal width (cm)','target']]
df1
```

```python
P_test=df1.iloc[149]
P_test
```

```python
df3=df1.drop(df1.index[149])
df3
```

```python
X=P_test.iloc[0]
Y=P_test.iloc[1]
(X,Y)
```

```python
list_Euclidean_Distance=[]
for col in range(len(df3)):
   list_Euclidean_Distance.append(math.sqrt( ((X-df3.iloc[col,0])**2)+((Y-df3.iloc[col,1])**2)))
list_Euclidean_Distance
```

```python
df2=pd.DataFrame(list_Euclidean_Distance)
df2.columns=['Euclidean Distance']
df2
```

```python
df2.min()
```
